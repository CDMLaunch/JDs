```python
df = pd.read_csv('df_description_processed.csv')
```

```python
# create the transform
vectorizer = CountVectorizer(stop_words = 'english')
# tokenize and build vocab
bag_of_words = vectorizer.fit_transform(df.clean_stemmed)
# summarize
sum_words = bag_of_words.sum(axis=0) 
words_freq = [(word, sum_words[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
words_freq[:10]
```

```python
df_counts = pd.DataFrame(dict(words_freq), index = ['frequency'])
df_counts = df_counts.T
df_counts.info()
```

```python

vectorizer = TfidfVectorizer(analyzer='word', ngram_range=(2, 4), max_df = .10, stop_words='english', norm = 'l1')
vz = vectorizer.fit_transform(df.clean_stemmed)

vz.shape
```

```python
tfidf = dict(zip(vectorizer.get_feature_names(), vectorizer.idf_))
tfidf = pd.DataFrame(columns=['tfidf']).from_dict(dict(tfidf), orient='index')
tfidf.columns = ['tfidf']
```

```python
tfidf.tfidf.hist(figsize=(15,7))
```

```python

def plot_word_cloud(terms):
    text = terms.index
    text = ' '.join(list(text))
    # lower max_font_size
    wordcloud = WordCloud(max_font_size=40).generate(text)
    plt.figure(figsize=(25, 25))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```

```python
plot_word_cloud(tfidf.sort_values(by=['tfidf'], ascending=True).head(40))
```

```python
plot_word_cloud(tfidf.sort_values(by=['tfidf'], ascending=False).head(100))
```
